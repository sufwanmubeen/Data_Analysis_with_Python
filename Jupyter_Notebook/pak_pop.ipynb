```python
# Total change in Urban Population
pop_2017_urban = df['ALL SEXES (URBAN)'].sum()
pop_1998_urban = df['POPULATION 1998 (URBAN)'].sum()
change_pop_urban = pop_2017_urban - pop_1998_urban
print("The change in urban population is: ",change_pop_urban)
```

```python
# Total change in Rural Population
pop_2017_rural = df['ALL SEXES (RURAL)'].sum()
pop_1998_rural = df['POPULATION 1998 (RURAL)'].sum()
change_pop_rural = pop_2017_rural - pop_1998_rural
print("The change in urban population is: ",change_pop_rural)
```

```python
# how to combine three columns in one new column i.e, male, female and transgender in Total Population column
df['Total Population (RURAL)'] = df[['MALE (RURAL)', 'FEMALE (RURAL)', 'TRANSGENDER (RURAL)']].sum(axis=1)
```

```python
df['Total Population (RURAL)'].sum()
```

```python
# how to combine three columns in one new column i.e, male, female and transgender in Total Population column
df['Total Population (URBAN)'] = df[['MALE (URBAN)', 'FEMALE (URBAN)', 'TRANSGENDER (URBAN)']].sum(axis=1)
```

```python
df.head()
```

```python
# how to combine two columns Rural and Urban Total population in Total Population column
df['Total Population'] = df[['Total Population (RURAL)', 'Total Population (URBAN)']].sum(axis=1)
```
