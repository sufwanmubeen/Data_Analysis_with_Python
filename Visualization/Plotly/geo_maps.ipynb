# Geo Maps in Python

```python
import plotly.express as px

df = px.data.election()
geojson = px.data.election_geojson()

print(df["district"][2])
print(geojson["features"][0]["properties"])
df.head()
```

```python
fig = px.choropleth_mapbox(df, geojson=geojson, color="winner",
                           locations="district", featureidkey="properties.district",
                           center={"lat": 45.5517, "lon": -73.7073},
                           mapbox_style="carto-positron", zoom=9)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

# Geomapping on COVID Data

```python
import pandas as pd
df = pd.read_json('covid_data.json')
df.head()
```

```python
df_cumulative=df.groupby(['country'])['cumulative_count'].mean().reset_index()
df_cumulative
```

```python
# geo map plotting
fig = px.choropleth(df_cumulative, color="cumulative_count",
                           locations="country",locationmode='country names',
                           title='Cases in countries of World',
                           range_color=[0,100000])
fig.show()
```

```python
fig = px.choropleth(df, color="cumulative_count",
                           locations="country",locationmode='country names',
                           title='Cases in countries of World',
                           range_color=[0,100000],
                           animation_frame='year_week')
fig.show()
```
